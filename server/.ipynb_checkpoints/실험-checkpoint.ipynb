```python
### step 0. 준비
import sys    # 시스템
import os     # 시스템

import pandas as pd    # 판다스 : 데이터분석 라이브러리
import numpy as np     # 넘파이 : 숫자, 행렬 데이터 라이브러리

from selenium import webdriver    # 웹 브라우저 자동화
import time                       # 시간 지연
import math                       # 올림, 내림
from selenium.webdriver.common.keys import Keys
from tqdm import tqdm   # 진행상황 표시
```

```python
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
options = Options()
options.add_argument('--start-maximized')
```

```python
# 스크롤 다운 함수
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0, 99999)")
    time.sleep(2)
```

```python
### step 1. url 수집

articles_number = int(input('크롤링 글 개수: '))

# 크롬 웹브라우저 실행
s=Service(ChromeDriverManager().install()) #크롬드라이버 설치 필요 없이 바로 가능
driver = webdriver.Chrome(service=s, chrome_options=options)


# 키워드로 접속
driver.get("https://www.g2b.go.kr/pt/menu/selectSubFrame.do?framesrc=/pt/menu/frameTgong.do?url=https://www.g2b.go.kr:8101/ep/tbid/tbidList.do?taskClCds=5&bidNm=%C1%B6%BB%E7&searchDtType=1&fromBidDt=2022/01/18&toBidDt=2022/02/17&fromOpenBidDt=&toOpenBidDt=&radOrgan=1&instNm=&area=&regYn=Y&bidSearchType=1&searchType=1")
time.sleep(4)

crawling_list = []
url_list = []
#fileDataList > div.result-list > ul > li:nth-child(1) > dl > dt > a > span.title
#fileDataList > div.result-list > ul > li:nth-child(1) > dl > dt > a
# url 수집
#crawling_list = driver.find_elements_by_tag_name('li > dl > dt > a')
crawling_list = driver.find_elements_by_tag_name('td > div > a')
#tr:nth-child(1) > td:nth-child(4) > div > a
#resultForm > div.results > table > tbody > tr:nth-child(1) > td:nth-child(4) > div > a

for crawling in crawling_list:
    url = crawling.get_attribute('href')   
    url_list.append(url)
    
driver.close()
url_list = url_list[:articles_number]
print('크롤링할 글 수: ', len(url_list))
url_list[0]
```

```python
### step 2. 크롤링

dict = {}
title = {}
count = 0
url = url_list[0]
driver = webdriver.Chrome(service=s, chrome_options=options)
driver.get(url)
time.sleep(1)

for url in tqdm(url_list):
    # url 열기
    driver = webdriver.Chrome(service=s,chrome_options=options)
    driver.get(url)
    time.sleep(2)
    
    try:
        title = driver.find_element_by_css_selector(' #contents > div.data-search-view > div.file-meta-table-pc > table > tbody > tr:nth-child(1) > td').text
    except:        
        title = driver.find_element_by_css_selector('#fileDetailTableArea > tbody > tr:nth-child(1) > td').text
    
    try:
        bun = driver.find_element_by_css_selector('#contents > div.data-search-view > div.file-meta-table-pc > table > tbody > tr:nth-child(2) > td:nth-child(2)').text
    except:
        bun = driver.find_element_by_css_selector('#fileDetailTableArea > tbody > tr:nth-child(2) > td:nth-child(2)').text
    
    try:
        file = driver.find_element_by_css_selector('#contents > div.data-search-view > div.file-meta-table-pc > table > tbody > tr:nth-child(7) > td:nth-child(2)').text
    except:
        file = driver.find_element_by_css_selector('#fileDetailTableArea > tbody > tr:nth-child(7) > td:nth-child(2)').text
        
    try:
        name = driver.find_element_by_css_selector('#contents > div.data-search-view > div.file-meta-table-pc > table > tbody > tr:nth-child(2) > td:nth-child(4)').text
    except:
        name = driver.find_element_by_css_selector('#fileDetailTableArea > tbody > tr:nth-child(2) > td:nth-child(4)').text
    try:
        control = driver.find_element_by_css_selector('#contents > div.data-search-view > div.file-meta-table-pc > table > tbody > tr:nth-child(3) > td:nth-child(2)').text
    except:
        control = driver.find_element_by_css_selector('#fileDetailTableArea > tbody > tr:nth-child(3) > td:nth-child(2)').text
    try:
        number = driver.find_element_by_css_selector('#telNo').text
    except:
        pass
   
    try:
        explain = driver.find_element_by_css_selector('#fileDetailTableArea > tbody > tr:nth-child(11) > td').text
    except :
        if driver.find_element_by_css_selector('#contents > div.data-search-view > div.file-meta-table-pc > table > tbody > tr:nth-child(11) > th').text=="URL":
            explain = driver.find_element_by_css_selector('#contents > div.data-search-view > div.file-meta-table-pc > table > tbody > tr:nth-child(12) > td').text
        else: 
            explain = driver.find_element_by_css_selector('#contents > div.data-search-view > div.file-meta-table-pc > table > tbody > tr:nth-child(11) > td').text

        
    
    
    target_info = {}
    target_info["제목"] = title
    target_info["분류"] = bun
    target_info["확장자"] = file
    target_info["제공기관"] = name
    target_info["관리부서"] = control
    target_info["관리부서 전화번호"] = number
    target_info["설명"] = explain
    
    driver.close()
    
    dict[count] = target_info
    count += 1
    time.sleep(1)
```

```python
# 판다스화
df = pd.DataFrame.from_dict(dict, 'index')
df
```
