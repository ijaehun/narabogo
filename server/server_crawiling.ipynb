```python
### step 0. 준비
import sys    # 시스템
import os     # 시스템

import pandas as pd    # 판다스 : 데이터분석 라이브러리
import numpy as np     # 넘파이 : 숫자, 행렬 데이터 라이브러리

from selenium import webdriver    # 웹 브라우저 자동화
import time                       # 시간 지연
import math                       # 올림, 내림
from selenium.webdriver.common.keys import Keys
from tqdm import tqdm   # 진행상황 표시
```

```python
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
options = Options()
options.add_argument('--start-maximized')

# 크롬 웹브라우저 실행
s=Service(ChromeDriverManager().install()) #크롬드라이버 설치 필요 없이 바로 가능
```

```python
url = ["https://www.g2b.go.kr/index.jsp"]
```

```python
url_list = ["https://www.g2b.go.kr/index.jsp"]
```

```python
### step 2. 크롤링

dict = {}
title = {}
count = 0
url = url_list[0]
driver = webdriver.Chrome(service=s, chrome_options=options)
driver.get(url)
time.sleep(1)

search = driver.find_element_by_css_selector('#searchForm > div > fieldset:nth-child(1) > ul > li:nth-child(4) > dl > dd.fr > a > strong')
search.click()
time.sleep(1)

driver.switch_to.frame('sub')
driver.switch_to.frame('left')

first = driver.find_element_by_xpath('//*[@id="000019"]/a')
first.click()
time.sleep(1)

second = driver.find_element_by_xpath('//*[@id="000022"]/a')
second.click()
time.sleep(1)

driver.switch_to.default_content()
driver.switch_to.frame('sub')
driver.switch_to.frame('main')

third = driver.find_element_by_xpath('//*[@id="bidNm"]')
third.send_keys('조사')

num1 = driver.find_element_by_xpath('//*[@id="fromBidDt"]')
num1.clear()
num1.send_keys('2021/07/01')

num2 = driver.find_element_by_xpath('//*[@id="toBidDt"]')
num2.clear()
num2.send_keys('2021/12/31')


search2 = driver.find_element_by_xpath('//*[@id="search"]/table/tbody/tr[7]/td/div/button[1]')
search2.send_keys(Keys.CONTROL + "\n")
time.sleep(1)

driver.switch_to.window(driver.window_handles[1])
time.sleep(1)

searchlist = driver.find_element_by_xpath('//*[@id="industrialCd"]')
searchlist.send_keys('1169')
time.sleep(1)

#searchlistclick= driver.find_element_by_css_selector('#bt_search > span')
searchlistclick = driver.find_element_by_xpath('//*[@id="bt_search"]')
searchlistclick.click()
time.sleep(1)

searchlistclick2 = driver.find_element_by_xpath('//*[@id="ebid"]/div[5]/table/tbody/tr/td[4]/a')
searchlistclick2.click()

driver.switch_to.window(driver.window_handles[0])
driver.switch_to.default_content()
driver.switch_to.frame('sub')
driver.switch_to.frame('main')

searchclick2 = driver.find_element_by_xpath('//*[@id="buttonwrap"]/div/a[1]/span')
searchclick2.click()


driver.switch_to.default_content()
driver.switch_to.frame('sub')
driver.switch_to.frame('main')
  
while True:
    try: 
        for i in range(1,31):
            title = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[1]/div').text
            bun = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[4]/div/a').text
            file = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[5]/div').text
            name = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[6]/div').text
            control = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[7]/div').text
            number = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[8]/div').text

            target_info = {}
            target_info["업무"] = title
            target_info["공고명"] = bun
            target_info["공고기관"] = file
            target_info["수요기관"] = name
            target_info["계약방법"] = control
            target_info["입력일시"] = number

            dict[count] = target_info
            count += 1
    except:
        break
                
    posting = driver.find_element_by_css_selector('#pagination > a.default')
    posting.click()
    time.sleep(3)
        
driver.close()


    

    
    
```
