```python
### step 2. 크롤링

dict = {}
title = {}
count = 0
url = url_list[0]
driver = webdriver.Chrome(service=s, chrome_options=options)
driver.get(url)
time.sleep(1)

for url in tqdm(url_list):
    
    driver = webdriver.Chrome(service=s,chrome_options=options)
    driver.get(url)
    time.sleep(2)

    driver.switch_to.frame('sub')
    driver.switch_to.frame('main')
    
    while True:
        try: 
            for i in range(1,11):
                title = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[1]/div').text
                bun = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[4]/div/a').text
                file = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[5]/div').text
                name = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[6]/div').text
                control = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[7]/div').text
                number = driver.find_element_by_xpath(f'//*[@id="resultForm"]/div[2]/table/tbody/tr[{i}]/td[8]/div').text

                target_info = {}
                target_info["업무"] = title
                target_info["공고명"] = bun
                target_info["공고기관"] = file
                target_info["수요기관"] = name
                target_info["계약방법"] = control
                target_info["입력일시"] = number

                dict[count] = target_info
                count += 1
        except:
            break
                
        posting = driver.find_element_by_css_selector('#pagination > a.default')
        posting.click()
        time.sleep(3)
        
    driver.close()

    

    
    
```
